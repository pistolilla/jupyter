# Using NLTK

```python
import nltk
#nltk.download('treebank')
from nltk.corpus import treebank
from nltk.probability import FreqDist, ConditionalFreqDist
```

```python
fd = FreqDist(treebank.words())
fd
```

```python
fd['the']
```

```python
cfd = ConditionalFreqDist(treebank.tagged_words())
cfd
```

```python
cfd['the']
```

```python
cfd['the'].max()
```

# Using collections

```python
import collections
counter = collections.Counter(treebank.words())
counter.most_common()
```
